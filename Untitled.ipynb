```python
#8. Iterar a través de los primeros cien enteros positivos, buscando los 
#múltiplos de 3 e imprimiendo y almacenándolos en
#una lista hasta encontrar los primeros 15 de ellos. Una vez encontrados, 
#continuar iterando en busca de los múltiplosde 4 y almacenarlos en otra lista.

def multiplo(valor, multiplo):
    """
    Funcion para calcular si el numero es multiplo
    utilizando el modulo de la division
    """
    resto = valor % multiplo
    if resto == 0:
        return True
    else:
        return False
 
# lista que contendra los valores multiplos
multiplos_3=[]
multiplos_4=[]
 
# bucle del 0 al 100
for i in range(0,45):
 
    if multiplo(i,3):
        multiplos_3.append(i)
for i in range(45,100):
    if multiplo(i,4):
        multiplos_4.append(i)
 
print ("Los multiplos de 3 son:", multiplos_3)
print ("")
print ("Los multiplos de 4 son:", multiplos_4)
```
